```python
import numpy as np
```

```python
X1 = np.array([1, 2, 3, 4, 5, 6])
Y = np.array([3000, 3250, 3500, 3750, 4000, 4250])
m = len(X1)
```

```python
X1 = X1.reshape(m,1)
print(X1)
print(X1.shape)
```

```python
pom = np.ones((m,1))
pom
```

```python
X = np.append(pom,X1, axis = 1)
X
```

```python
np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, Y))
```

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
regression.fit(X1, Y)
#a
print(regression.coef_[0])
#b
print(regression.intercept_)
```

```python
import pandas as pd
import plotly.express as px
```

```python
X1 = np.array([1,2,3,4,5,6])
Y = np.array([3000, 3250, 3500, 3750, 4000, 4250])
m = len(X1)
```

```python
print(f'Lata pracy: {X1}')
print(f'Wynagrodzenie: {Y}')
print(f'Liczba próbek: {m}')
```

```python
X1 = X1.reshape(m,1)
X1
```

```python
Y = Y.reshape(-1,1)
print(Y)
```

```python
bias = np.ones((m,1))
print(bias)
print(bias.shape)
```

```python
X = np.append(bias,X1,axis=1)
X
```

```python
eta = 0.01

w = np.random.rand(2,1)
w
```

```python
intercept = []
coef = []

for i in range(3000):
  gradient = (2/m) * X.T.dot(X.dot(w) - Y)
  w = w - eta * gradient
  intercept.append(w[0][0])
  coef.append(w[1][0])

print(w)
```

```python
df = pd.DataFrame(data = {'intercept': intercept, 'coef': coef})
df.head(10)
```

```python
df.tail(10)
```

```python
px.line(df, y='intercept', width=800, title='Dopasowanie: intercept')
```

```python
px.line(df, y='coef', width=800, title='Dopasowanie: coef')
```
